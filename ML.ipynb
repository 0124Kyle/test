```python
# 匯入資料庫+欄位顯示設定

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',30)
```

```python
# 匯入資料

data = pd.read_csv('heart_attack_prediction_dataset.csv')
data.head()
```

```python
# 資料觀察

data.info()
```

```python
# 資料描述

data.describe()
```

```python
# 資料規模

print('rows:共',str(data.shape[0]),'筆')
print('columns:共',str(data.shape[1]),'筆')
```

```python
# 分離血壓數值(高、低血壓)

hpressure = np.array([])
lpressure = np.array([])
for i in range(len(data['Blood Pressure'])):
    splits = data['Blood Pressure'][i].split('/')
    hpressure = np.append(hpressure,splits[0])
    lpressure = np.append(lpressure,splits[1])

data['High blood pressure'] = hpressure
data['Low blood pressure'] = lpressure
data.drop(columns='Blood Pressure',inplace=True)
```

```python
#查看有無空值(缺失值) 

data.isna().sum()
```

```python
# 列出資料欄位

data.columns

# Cholestrol:膽固醇
# Heart Rate:心率
# Diabetes:糖尿病
# Family History:家族病史
# Smoking:是否吸菸
# Obesity:肥胖
# Alcohol Consumption:酒精攝取
# Exercise Hours Per Week:每周運動時數
# Diet:飲食習慣
# Previous Heart Problems:先前是否有心臟病史
# Medication Use:藥物使用
# Stress Level:壓力等級
# Sedentary Hours Per Day:每日久坐時間
# Income:收入
# BMI:身體指數
# Teiglycerides:三酸甘油酯
# Physical Activity Days Per Week:每周運動天數
# Sleep Hours Per Day:每天睡眠小時數
# Country:國家
# Continent:大陸
# Hemisphere:半球
# Heart Attack Risk:心臟病風險
# High blood pressure:高血壓
# Low blood pressure:低血壓 
```

```python
# 因為把原本血壓的欄位切割成高低血壓，要把資料型態改為整數

data['High blood pressure'] = data['High blood pressure'].astype(int)
data['Low blood pressure'] = data['Low blood pressure'].astype(int)
```

```python
data.info()
```

```python
# 改變欄位順序

data = data.reindex(columns=['Patient ID', 'Age', 'Sex', 'Cholesterol', 'Heart Rate', 'High blood pressure', 'Low blood pressure', 'Diabetes',
       'Family History', 'Smoking', 'Obesity', 'Alcohol Consumption',
       'Exercise Hours Per Week', 'Diet', 'Previous Heart Problems',
       'Medication Use', 'Stress Level', 'Sedentary Hours Per Day', 'Income',
       'BMI', 'Triglycerides', 'Physical Activity Days Per Week',
       'Sleep Hours Per Day', 'Country', 'Continent', 'Hemisphere',
       'Heart Attack Risk'])

data.head()
```

# 1. 先使用膽固醇跟糖尿病來觀察是否與心臟病風險有關

```python
# 提取欄位資料+分割資料集

from sklearn.model_selection import train_test_split

df1 = pd.DataFrame({})
df1['Cholesterol'] = data['Cholesterol']
df1['Diabetes'] = data['Diabetes']
df1['Heart Attack Risk'] = data['Heart Attack Risk']


X = df1.iloc[:,:-1].values
y = df1.iloc[:,-1].values

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# 進行標準化轉換
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# 利用邏輯迴歸演算法進行模型訓練

from sklearn.linear_model import LogisticRegression

regressor = LogisticRegression()
regressor.fit(X_train,y_train)
```

```python
# 利用測試集生成模型預測結果

y_pred = regressor.predict(X_test)
y_pred
```

```python
# 利用模型預測結果和標準答案進行回測並計算準確度

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred,y_test)
print(cm)
print()
print('Accuracy Score:',round(accuracy_score(y_pred,y_test),2)*100,'%',sep='')
```

# 2. 使用家族病史跟先前病史觀察是否跟心臟病風險有關

```python
# 提取欄位資料+分割資料集

df2 = pd.DataFrame({})

df2['Family History'] = data['Family History']
df2['Previous Heart Problem'] = data['Previous Heart Problems']
df2['Heart Attack Risk'] = data['Heart Attack Risk']

df2
```

```python
X1 = df2.iloc[:,:-1].values
y1 = df2.iloc[:,-1].values

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dc = DecisionTreeClassifier(criterion='entropy',random_state=0)
dc.fit(X1_train,y1_train)
```

```python
y1_pred = dc.predict(X1_test)
y1_pred
```

```python
cm1 = confusion_matrix(y1_pred,y1_test)
print(cm1)
print()
print('Accuracy Score:',round(accuracy_score(y1_pred,y1_test),2)*100,'%',sep='')
```
